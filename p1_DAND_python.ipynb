```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn

df=pd.read_csv("stroopdata.csv")

df['Congruent'].hist(bins='auto', label='Congruent')
df['Incongruent'].hist(bins='auto', label='Incongruent')
plt.title("Frequency Histogram")
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.legend(loc='upper right')

plt.show()

```

```python
df.boxplot()
plt.show()
```
